# Image to Pencil Sketch ML Project

Approach:

1.  We first convert the image into a grayscale image
2.  We invert the grayscale image
3.  Apply Gaussian Blur to the inverted grayscale image
4.  Invert the blurred image
5.  Finally convert the blurred image to pencil sketch

We use gaussian blur as it smoothens the image and removes outlier pixels and noise in the image



## Importing Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

## Code

```python
image = cv2.imread("animal.jpg")
cv2_imshow(image)
```

```python
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayImage)
```

```python
invertedImage = 255 - grayImage
cv2_imshow(invertedImage)
```

```python
blurredImage = cv2.GaussianBlur(invertedImage, (21, 21), 0)
cv2_imshow(blurredImage)
```

```python
invertedBlurredImage = 255 - blurredImage
cv2_imshow(invertedBlurredImage)
```

```python
pencil_sketch = cv2.divide(grayImage, invertedBlurredImage, scale=256.0)
cv2_imshow(pencil_sketch)
```

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
